```python
import numpy as np
import matplotlib.pyplot as plt

# .npz 파일 로드
# file_path = '/Users/songhune/Library/Mobile Documents/com~apple~CloudDocs/Workspace/moving-mnist-train.npz'
file_path = '/Users/songhune/Downloads/data/kth/kth_train_fold_0.npz'
#file_path = '/Users/songhune/Downloads/data/climate.npz'


data = np.load(file_path)

# 각 배열의 이름과 shape 출력
for key in data.files:
    print(f"Array name: {key}")
    print(f"Array shape: {data[key].shape}")
    print()

# 데이터 요약 함수
def summarize_data(data):
    for key in data.files:
        arr = data[key]
        print(f"{key} 요약:")
        print(f"형태: {arr.shape}")
        print(f"평균: {np.mean(arr)}")
        print(f"표준편차: {np.std(arr)}")
        print()

# 데이터 시각화 함수
def visualize_data(data, array_name):
    if array_name not in data.files:
        print(f"{array_name} 배열이 데이터셋에 없습니다.")
        return
    
    array_data = data[array_name]
    fig, axs = plt.subplots(1, 5, figsize=(15, 3))
    for i in range(5):
        axs[i].imshow(array_data[i, 0], cmap='gray')
        axs[i].axis('off')
    plt.show()

# 각 배열의 이름과 shape 출력
summarize_data(data)

# input_raw_data 데이터 시각화
visualize_data(data, 'input_raw_data')
#visualize_data(data,'temp') // 3차원 데이터는 시각화가 안됨
```

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio.v3 as iio
# .npz 파일 로드
file_path = '/Users/songhune/Library/Mobile Documents/com~apple~CloudDocs/Workspace/moving-mnist-train.npz'
data = np.load(file_path)

# 각 배열의 이름과 shape 출력
def summarize_data(data):
    for key in data.files:
        arr = data[key]
        print(f"{key} 요약:")
        print(f"형태: {arr.shape}")
        print(f"평균: {np.mean(arr)}")
        print(f"표준편차: {np.std(arr)}")
        print()

# 데이터 시각화 함수
def visualize_data(data, array_name):
    if array_name not in data.files:
        print(f"{array_name} 배열이 데이터셋에 없습니다.")
        return
    
    array_data = data[array_name]
    fig, axs = plt.subplots(1, 5, figsize=(15, 3))
    for i in range(5):
        axs[i].imshow(array_data[i, 0], cmap='gray')
        axs[i].axis('off')
    plt.show()

# GIF 생성 함수
def create_gif(data, array_name, gif_name='animation.gif'):
    if array_name not in data.files:
        print(f"{array_name} 배열이 데이터셋에 없습니다.")
        return
    
    array_data = data[array_name]
    frames = []
    
    for i in range(array_data.shape[0]):
        fig, ax = plt.subplots()
        ax.imshow(array_data[i, 0], cmap='gray')
        ax.axis('off')
        fig.canvas.draw()
        
        image = np.frombuffer(fig.canvas.tostring_rgb(), dtype='uint8')
        image = image.reshape(fig.canvas.get_width_height()[::-1] + (3,))
        frames.append(image)
        plt.close(fig)
    
    iio.mimsave(gif_name, frames, fps=5)
    print(f"GIF 생성 완료: {gif_name}")

# 각 배열의 이름과 shape 출력
summarize_data(data)

# input_raw_data 데이터 시각화
visualize_data(data, 'input_raw_data')

# input_raw_data를 동영상처럼 보여주는 gif 생성
create_gif(data, 'input_raw_data')
```
